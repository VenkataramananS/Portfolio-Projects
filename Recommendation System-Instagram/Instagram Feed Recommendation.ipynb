```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import text
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('C:\\projects\\recommendation system\\Instagram_feed_recommendation\\ig_recommendation.csv')
```

```python
df.head()
```

```python
# Choosing Captions for recommendation system
captions = df.Caption.to_list()
# Cleaning the captions by removing the stopwords
caption_tfidf = text.TfidfVectorizer(input = captions, stop_words="english")
# Converting the vectorized captions into a matrix to build a recommendor system based on cosine similarity method
vectorized_matrix = caption_tfidf.fit_transform(captions)
caption_similarity = cosine_similarity(vectorized_matrix)
```

```python
def function_to_rephrase(index):
    return ",".join(df.Caption.loc[index.argsort()])
```

```python
df['Recommendation'] = [function_to_rephrase(i) for i in caption_similarity]
```
